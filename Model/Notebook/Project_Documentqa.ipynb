```python
import re
import pandas as pd
```

```python
print("Starting data preprocessing...")
def clean_text(text):
    """
    Cleans a given text by lowercasing, removing special characters,
    and normalizing whitespace.
    """
    if not isinstance(text, str):
        return ""
    # Lowercase the text
    text = text.lower()
    # Remove anything that is not a letter or a space
    text = re.sub(r'[^a-z\s]', '', text)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
print("\nCleaning the legal summarization dataset...")

# We use .copy() to avoid SettingWithCopyWarning in pandas.
cleaned_summary_df = legal_summary_df.copy()

# Print the actual columns to confirm their names.
print(f"Available columns in the DataFrame: {cleaned_summary_df.columns.tolist()}")

# ERROR FIX: Using the correct column names 'judgement' and 'EN_summary'
# as identified from your error output.
cleaned_summary_df['EN_judgment_cleaned'] = cleaned_summary_df['judgement'].apply(clean_text)
cleaned_summary_df['EN_summary_cleaned'] = cleaned_summary_df['EN_summary'].apply(clean_text)

print("Summarization dataset cleaned successfully!")
print("\n--- Sample of Cleaned Summarization Data ---")
print(cleaned_summary_df[['EN_judgment_cleaned', 'EN_summary_cleaned']].head())
print("--------------------------------------------\n")
```

```python
model_checkpoint = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
print(f"Tokenizer for '{model_checkpoint}' loaded successfully.")
```

```python
max_input_length = 1024  # Max length for the input document
max_target_length = 128 # Max length for the output summary

def preprocess_function(examples):
    # Tokenize the input documents
    model_inputs = tokenizer(
        examples["EN_judgment_cleaned"],
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )

    # Tokenize the target summaries
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["EN_summary_cleaned"],
            max_length=max_target_length,
            truncation=True,
            padding="max_length"
        )

    # The model needs the tokenized summaries under the 'labels' key
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
train_df, val_df = train_test_split(df, test_size=0.1, random_state=42)

# Convert pandas DataFrames to Hugging Face Dataset objects
train_dataset_raw = Dataset.from_pandas(train_df)
val_dataset_raw = Dataset.from_pandas(val_df)

print(f"\nData split into {len(train_dataset_raw)} training samples and {len(val_dataset_raw)} validation samples.")
```

```python
print("Applying tokenization to the datasets...")
tokenized_train_dataset = train_dataset_raw.map(preprocess_function, batched=True)
tokenized_val_dataset = val_dataset_raw.map(preprocess_function, batched=True)
print("Tokenization complete!")
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_checkpoint)

print("\nSetup complete. The data is now ready for the training phase.")
```

```python
from transformers import AutoTokenizer, T5ForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer, DataCollatorForSeq2Seq
from datasets import Dataset
import pandas as pd
import os
import numpy as np
import evaluate

print("Setting up to train the T5 model...")
```

```python
drive_path = '/content/drive/My Drive/Legal_AI_Project/'
cleaned_file_path = os.path.join(drive_path, 'cleaned_legal_summarization_data.csv')
df = pd.read_csv(cleaned_file_path)
df.dropna(subset=['EN_judgment_cleaned', 'EN_summary_cleaned'], inplace=True)
print("Cleaned dataset loaded successfully.")
```

```python
model_checkpoint_t5 = "google-t5/t5-base"
tokenizer_t5 = AutoTokenizer.from_pretrained(model_checkpoint_t5)
print(f"Tokenizer for '{model_checkpoint_t5}' loaded successfully.")
```

```python
prefix = "summarize: "
max_input_length = 1024
max_target_length = 256

def preprocess_function_t5(examples):
    # Add the prefix to the input documents
    inputs = [prefix + doc for doc in examples["EN_judgment_cleaned"]]
    model_inputs = tokenizer_t5(
        inputs,
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )

    # Tokenize the target summaries
    labels = tokenizer_t5(
        text_target=examples["EN_summary_cleaned"],
        max_length=max_target_length,
        truncation=True,
        padding="max_length"
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Convert DataFrame to Dataset object
t5_dataset = Dataset.from_pandas(df)

print("\nApplying T5 tokenization to the dataset...")
tokenized_dataset_t5 = t5_dataset.map(preprocess_function_t5, batched=True, remove_columns=df.columns.tolist())

# Split into train and validation sets
train_val_split_t5 = tokenized_dataset_t5.train_test_split(test_size=0.1, seed=42)
train_dataset_t5 = train_val_split_t5['train']
val_dataset_t5 = train_val_split_t5['test']
print("Tokenization and data splitting complete for T5.")
```

```python
rouge_metric = evaluate.load("rouge")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer_t5.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer_t5.pad_token_id)
    decoded_labels = tokenizer_t5.batch_decode(labels, skip_special_tokens=True)
    result = rouge_metric.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)
    result = {key: value * 100 for key, value in result.items()}
    return {k: round(v, 4) for k, v in result.items()}

# Load a fresh T5 model
model_t5 = T5ForConditionalGeneration.from_pretrained(model_checkpoint_t5)

# Define new output directory and training arguments
output_dir_t5 = os.path.join(drive_path, "training_checkpoints_t5")

training_args_t5 = Seq2SeqTrainingArguments(
    output_dir=output_dir_t5,
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4, # T5 is more memory efficient than PEGASUS
    per_device_eval_batch_size=4,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=3,
    predict_with_generate=True,
    logging_steps=100,
    save_strategy="epoch",
)

data_collator_t5 = DataCollatorForSeq2Seq(tokenizer=tokenizer_t5, model=model_t5)

trainer_t5 = Seq2SeqTrainer(
    model=model_t5,
    args=training_args_t5,
    train_dataset=train_dataset_t5,
    eval_dataset=val_dataset_t5,
    data_collator=data_collator_t5,
    compute_metrics=compute_metrics,
)

print("\nT5 trainer configured. Starting the new training process...")
```

```python
trainer_t5.train()

print("\nT5 training complete!")
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import pandas as pd
import os

print("Starting the QA Knowledge Base creation...")
```

```python
drive_path = '/content/drive/My Drive/Legal_AI_Project/'
cleaned_file_path = os.path.join(drive_path, 'cleaned_legal_summarization_data.csv')
df = pd.read_csv(cleaned_file_path)
df.dropna(subset=['EN_judgment_cleaned'], inplace=True)

# For this step, we only need the documents themselves.
documents = df['EN_judgment_cleaned'].tolist()
print(f"Loaded {len(documents)} cleaned legal documents.")
```

```python
print("\nLoading the Sentence Transformer model...")
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Embedding model loaded successfully.")
```

```python
print("\nGenerating embeddings for all documents... (This may take a while)")
document_embeddings = embedding_model.encode(documents, show_progress_bar=True)
print("Embeddings generated successfully.")
print(f"Shape of the embedding matrix: {document_embeddings.shape}")
```

```python
print("\nBuilding the FAISS index...")

# Get the dimension of our embeddings
d = document_embeddings.shape[1]

# We'll use a simple IndexIVFFlat index for a good balance of speed and accuracy
# First, we need a quantizer index (a flat L2 index in this case)
quantizer = faiss.IndexFlatL2(d)
# Then we create the IVF index
nlist = 100 # How many cells to partition the index into
index = faiss.IndexIVFFlat(quantizer, d, nlist, faiss.METRIC_L2)

# The index needs to be trained on the data to learn the partitions
index.train(document_embeddings)
# Add the actual vectors to the index
index.add(document_embeddings)

print(f"FAISS index built successfully. Total vectors in index: {index.ntotal}")
```

```python
print("\nSaving the FAISS index and documents to Google Drive...")

# Define paths
faiss_index_path = os.path.join(drive_path, 'legal_documents.index')
documents_path = os.path.join(drive_path, 'legal_documents.pkl') # Using pickle to save the list

# Save the FAISS index
faiss.write_index(index, faiss_index_path)

# Save the documents list using pickle
import pickle
with open(documents_path, 'wb') as f:
    pickle.dump(documents, f)

print(f"FAISS index saved to: {faiss_index_path}")
print(f"Documents list saved to: {documents_path}")
print("\nQA Knowledge Base setup is complete!")
```

# QA

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sentence_transformers import SentenceTransformer
import faiss
import pickle
import os
import torch

print("Setting up the QA Pipeline...")
```

```python
drive_path = '/content/drive/My Drive/Legal_AI_Project/'
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the fine-tuned summarization model (our "Generator")
# FINAL FIX: Pointing to the last known successfully saved and high-performing model.
model_path = os.path.join(drive_path, 'final_summarization_model_v2')
print(f"\nLoading fine-tuned model from: {model_path}")
qa_model = AutoModelForSeq2SeqLM.from_pretrained(model_path).to(device)
qa_tokenizer = AutoTokenizer.from_pretrained(model_path)

# Load the embedding model (our "Retriever")
print("Loading the Sentence Transformer model...")
embedding_model = SentenceTransformer('all-MiniLM-L6-v2', device=device)

# Load the FAISS index and the documents list
faiss_index_path = os.path.join(drive_path, 'legal_documents.index')
documents_path = os.path.join(drive_path, 'legal_documents.pkl')

print("Loading the FAISS index and documents...")
faiss_index = faiss.read_index(faiss_index_path)
with open(documents_path, 'rb') as f:
    documents = pickle.load(f)

print("All components loaded successfully!")
```

```python
def answer_question(question, top_k=5):
    """
    This function performs the full RAG pipeline:
    1. Encodes the user's question into an embedding.
    2. Searches the FAISS index for the most relevant documents.
    3. Combines the question and the retrieved documents into a prompt.
    4. Uses the fine-tuned model to generate a final answer.
    """
    print(f"\nSearching for documents related to: '{question}'")
    # 1. Retrieval Stage
    question_embedding = embedding_model.encode([question])
    distances, indices = faiss_index.search(question_embedding, top_k)

    # Get the actual text of the most relevant documents
    retrieved_docs = [documents[i] for i in indices[0]]
    context = " ".join(retrieved_docs)
    print("Relevant documents retrieved successfully.")

    # 2. Generation Stage
    # We create a prompt that instructs the model how to behave.
    prompt = f"""
    Answer the following question based only on the provided context.
    If the context does not contain the answer, say "I cannot find the answer in the provided documents."

    Context:
    {context}

    Question:
    {question}

    Answer:
    """

    print("Generating an answer with the fine-tuned model...")
    inputs = qa_tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).to(device)
    summary_ids = qa_model.generate(
        inputs.input_ids,
        num_beams=4,
        max_length=256, # Max length for the answer
        early_stopping=True
    )
    answer = qa_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return answer
```

```python
sample_question = "what is the law regarding bail"

final_answer = answer_question(sample_question)

print("\n" + "="*25)
print(f"QUESTION: {sample_question}")
print(f"ANSWER: {final_answer}")
print("="*25)
```
